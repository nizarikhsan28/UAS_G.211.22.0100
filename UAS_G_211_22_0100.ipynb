```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
data_curah_hujan = pd.read_csv('data_curah_hujan.csv', sep=';')
data_suhu_permukaan = pd.read_csv('data.csv')
```

**1.**

```python
# Plotting
plt.figure(figsize=(10, 5))

# Mengatur lebar bar
bar_width = 0.20

# Mengatur posisi bar untuk setiap periode curah hujan
bulan = np.arange(len(data_curah_hujan))
posisi_1981_1990 = bulan
posisi_1991_2000 = bulan + bar_width
posisi_2001_2010 = bulan + 2 * bar_width

# Menambahkan gridlines horizontal
ytick_values = np.arange(0, 720, 20)
for y in ytick_values:
    linewidth = 1.5 if y % 100 == 0 else 0.8  # Set linewidth to 1 for values divisible by 100, 0.5 for others
    plt.axhline(y, color='gray', linestyle='-', linewidth=linewidth)

# Bar plot untuk curah hujan tiap periode
plt.bar(posisi_1981_1990, data_curah_hujan['Curah Hujan 1981-1990'], color='blue', width=bar_width, label='1981-1990', zorder=2)
plt.bar(posisi_1991_2000, data_curah_hujan['Curah Hujan 1991-2000'], color='red', width=bar_width, label='1991-2000', zorder=2)
plt.bar(posisi_2001_2010, data_curah_hujan['Curah Hujan 2001-2010'], color='green', width=bar_width, label='2001-2010', zorder=2)

# Menampilkan legenda menyamping di tengah
plt.legend(bbox_to_anchor=(0.5, 1.05), loc='upper center', ncol=3, frameon=False, prop={'size': 12}, markerscale=8)

# Konfigurasi plot
plt.xlabel('Bulan')
plt.ylabel('Curah Hujan (mm)')
plt.xticks(bulan + bar_width, data_curah_hujan['Bulan'])

# Menghapus spines atas dan kanan
plt.box(on=True)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Tampilkan plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 4))

# Garis plot untuk setiap jenis data
plt.plot(range(len(data_suhu_permukaan)), data_suhu_permukaan['asimilasi'], marker='o', color='blue', label='Asimilasi')
plt.plot(range(len(data_suhu_permukaan)), data_suhu_permukaan['observasi'], marker='o', color='orange', label='Observasi')
plt.plot(range(len(data_suhu_permukaan)), data_suhu_permukaan['non_asimilasi'], marker='o', color='red', label='Non Asimilasi')

# Mengatur plot
plt.xlabel('Waktu', labelpad=20)  # Mengatur jarak antara label x-axis dan sumbu x
plt.ylabel('Celcius')
plt.title('Suhu Permukaan')
plt.xticks(range(len(data_suhu_permukaan)), data_suhu_permukaan['waktu'])
plt.ylim(20, 27)

# Memindahkan legenda ke sebelah kiri dan mengatur ukuran font
plt.legend(loc='center left', frameon=False, bbox_to_anchor=(1, 0.37), fontsize='11')

# Menambahkan garis grid horizontal
plt.grid(axis='y', linestyle='-', linewidth=0.50)

# Menambahkan garis horizontal pada y=20
plt.axhline(y=20, color='gray', linestyle='-', xmin=0, xmax=len(data_suhu_permukaan)-1)

# Menghilangkan tepi atas dan kanan
plt.box(False)

# Mengatur label x-axis
plt.xticks(range(len(data_suhu_permukaan)), data_suhu_permukaan['waktu'])

# Mencari posisi indeks untuk '00.00 UTC' dan '21.00 UTC'
pos_00 = data_suhu_permukaan[data_suhu_permukaan['waktu'] == '00.00 UTC'].index[0]
pos_21 = data_suhu_permukaan[data_suhu_permukaan['waktu'] == '21.00 UTC'].index[0]

# Menghitung posisi untuk label '24 juli 2013'
pos_24 = (pos_00 + pos_21) / 2

# Menambahkan label '24 juli 2013' di bawah dan di tengah antara '00.00 UTC' dan '21.00 UTC'
plt.text(pos_24, 19, '24 Juli 2013', ha='center')

# Mencari posisi indeks untuk '00.00 UTC' di hari berikutnya
pos_25 = data_suhu_permukaan[data_suhu_permukaan['tanggal'] == '25 juli 2013'].index[0]

# Menambahkan label '25 juli 2013' di bawah '00.00 UTC' di hari berikutnya
plt.text(pos_25, 19, '25 Juli 2013', ha='center')

# Menampilkan plot
plt.show()
```
